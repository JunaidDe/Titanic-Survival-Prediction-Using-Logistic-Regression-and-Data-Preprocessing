```python
import pandas as pd

# Define the file paths
train_file_path = '/content/drive/My Drive/archive (4)/train.csv'
test_file_path = '/content/drive/My Drive/archive (4)/test.csv'

# Load the datasets into DataFrame
train_data = pd.read_csv(train_file_path)
test_data = pd.read_csv(test_file_path)

# Display the first few rows of the datasets
train_data.head(), test_data.head()
```

```python
# Check for missing values in both datasets
train_missing = train_data.isnull().sum()
test_missing = test_data.isnull().sum()

# Display missing values count
print("Train Data Missing Values:\n", train_missing)
print("\nTest Data Missing Values:\n", test_missing)

# Fill or drop missing values (choose based on the column)
train_data.fillna(train_data.mean(), inplace=True)  # Filling missing numerical values with the mean
test_data.fillna(test_data.mean(), inplace=True)  # Same for test data
```

```python
# Check data types of columns
print(train_data.dtypes)
print(test_data.dtypes)
```

```python
# Check if the columns meant to be numeric have any non-numeric values
train_data['Age'] = pd.to_numeric(train_data['Age'], errors='coerce')
train_data['Fare'] = pd.to_numeric(train_data['Fare'], errors='coerce')

# Verify if there are any non-numeric values after conversion
print(train_data['Age'].isnull().sum())
print(train_data['Fare'].isnull().sum())
```

```python
# Fill missing values with the mean of the column
train_data['Age'].fillna(train_data['Age'].mean(), inplace=True)
train_data['Fare'].fillna(train_data['Fare'].mean(), inplace=True)
```

```python
# Check column names in both train and test data
print("Train Data Columns:", train_data.columns)
print("Test Data Columns:", test_data.columns)
```

```python
# Split train data into features (X) and target (y)
X_train = train_data.drop(columns=['Survived', 'PassengerId', 'Name', 'Ticket', 'Cabin'])
y_train = train_data['Survived']

# Split test data into features (X)
X_test = test_data.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])
```

```python
from sklearn.linear_model import LogisticRegression

# Create a Logistic Regression model
log_reg_model = LogisticRegression(max_iter=200)

# Train the model on the training data
log_reg_model.fit(X_train, y_train)
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer object to replace missing values with the median
imputer = SimpleImputer(strategy='median')

# Apply imputer to both training and testing datasets
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

```

```python
# Drop rows with missing values in both training and testing data
X_train_dropped = X_train.dropna()
y_train_dropped = y_train[X_train_dropped.index]
X_test_dropped = X_test.dropna()
```

```python
# Re-train the Logistic Regression model
log_reg_model.fit(X_train_imputed, y_train)  # If you used imputation
# or
log_reg_model.fit(X_train_dropped, y_train_dropped)  # If you dropped rows with missing values
```

```python
# Predict on the test data
y_pred = log_reg_model.predict(X_test_imputed)  # Use imputed data for prediction

# Save the predictions to a DataFrame
predictions = pd.DataFrame({'PassengerId': test_data['PassengerId'], 'Survived': y_pred})

# Save predictions to a CSV file
predictions.to_csv('/content/drive/My Drive/archive (4)/titanic_predictions.csv', index=False)

print("Predictions saved successfully!")
```

```python
from sklearn.model_selection import cross_val_score

# Evaluate the model using cross-validation
cross_val_accuracy = cross_val_score(log_reg_model, X_train_imputed, y_train, cv=5, scoring='accuracy')

print(f"Cross-validation Accuracy: {cross_val_accuracy.mean():.4f}")
```
